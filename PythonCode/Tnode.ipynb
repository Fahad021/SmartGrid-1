```python
'''
[LICENSE]
Copyright (c) 2016, Ahmed Chaouachi
IREQ: Institut de Recherche d'Hydro Qu√©bec
All rights reserved.
[\LICENSE]
'''
#self.type = {'PQ':0, 'PQV':1, 'PV':2, 'SWING':3}
class Tnode:
    '''
    The name of this class: T for Transmission , node for node
    This is a mother class of the classes TnodePQ, TnodePV, TnodePQV, TnodeSWING
    
    Attributes:
        cdfNum      - number in the cdf file
        busType     -  [1]PQ ,  [2]PV, [3]SWING, [4]PQV
        busName     - name
        busArea     - geographic Area, the areas are separable
        busSubArea  - geographic subArea
        busZone     - geographic zone
        busLoadMW   - P^d : active load
        busLoadMVAR - Q^d : reactive load
        busGL       - GL:  real component of shunt admittance to ground.
        busBL       - BL:  imaginary component of shunt admittance to ground.
        busKV       - V: base kV
        --------------------------
        desVoltage  - desired voltage
        --------------------------
        busVarAngle       - voltage magnitude
        busVarVoltage     - voltage magnitude
        busVarGenMW       - active generation (MW)
        busVarGenMVAR     - reactive generation (MVAR)

    '''
    """ NUMBER OF NODES IN THE TNETWROK """
    N_SIZE    =0
    PQ_SIZE   =0
    PV_SIZE   =0
    PQV_SIZE  =0
    SWING_SIZE=0   
    #def __init__(self, cdfNum, busType, busName, 
    #             busArea, busSubArea, busZone, busLoadMW, busLoadMVAR, busGL, busBL,busKV,
    #             busVarAngle,busVarVoltage, busVarGenMW, busVarGenMVAR, desVoltage):
    def __init__(self, params):
        #""" THE IEEE CDF PROPERTIES: BUS DATA """
        if 'cdfNum' in params.keys() :self.cdfNum = params['cdfNum'] 
        if 'busType' in params.keys() :self.busType   = params['busType']
        if 'busName' in params.keys() :self.busName   = params['busName']
        if 'busArea' in params.keys() :self.busArea   = params['busArea']
        if 'busSubArea' in params.keys() :self.busSubArea= params['busSubArea']
        if 'busZone' in params.keys() :self.busZone   = params['busZone']
        if 'busLoadMW' in params.keys() :self.busLoadMW = params['busLoadMW']
        if 'busLoadMVAR' in params.keys() :self.busLoadMVAR = params['busLoadMVAR']
        if 'busGL' in params.keys() :self.busGL  = params['busGL']
        if 'busBL' in params.keys() :self.busBL  = params['busBL']
        if 'busKV' in params.keys() :self.baseKV = params['busKV']
        # Desired Voltage
        if 'desVoltage' in params.keys() :self.desVoltage = params['desVoltage']
        #""" OTHER PROPERTIES - variables """
        if 'busVarAngle' in params.keys() :self.busVarAngle = params['busVarAngle']
        if 'busVarVoltage' in params.keys() :self.busVarVoltage = params['busVarVoltage']
        if 'busVarGenMW' in params.keys() :self.busVarGenMW = params['busVarGenMW']
        if 'busVarGenMVAR' in params.keys() :self.busVarGenMVAR =params['busVarGenMVAR']
        
        
        #""" Update the nodes' number """
        Tnode.N_SIZE+=1
        if self.busType == 1: Tnode.PQ_SIZE +=1
        elif self.busType == 2: Tnode.PV_SIZE+=1
        elif self.busType == 3: Tnode.SWING_SIZE +=1
        elif self.busType == 4: Tnode.PQV_SIZE+=1
        
        
            
        
        
        
        
```
